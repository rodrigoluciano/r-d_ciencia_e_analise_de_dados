```python
dict_fillna = {'price': ['100', 'KLD100', 400, 'ADL100']}
df_fillna = pd.DataFrame(dict_fillna)
df_fillna
```

```python
df_fillna['price'] = pd.to_numeric(df_fillna['price'], errors='coerce')
df_fillna
```

```python
df_fillna['price'] = df_fillna['price'].fillna(0)
```

```python
df_fillna
```

```python
dict_replace = {'price': ['100', 'KLD100', 400, 'ADL100']}
df_replace = pd.DataFrame(dict_replace)
df_replace['price'] = pd.to_numeric(df_replace['price'], errors='coerce')
df_replace
```

A função dropna() é utilizada para remover linhas ou colunas com valores Null/None/NA e retorna um novo DataFrame, podendo criar valores null usando None e utilizando propriedades de numpy.nan. Syntax: df.dropna(self, axis=0, 'how'='any', thresh=None, subset=None, inplace=False)

1 - axis: Os valores possíveis são {0 ou 'index', 1 ou 'columns'}, default value is 0 se o valor for 0 irá 'dropar' linhas com valores null. Caso o valor seja 1 irá dropar columns

2 - how: Os valores possíveis são {'any','all'}, por padrão 'any' dropa a linha/coluna caso os valores seja null, se for 'all' dropa todos os valores perdidos

3 - thresh: caso um valor int para especificar o threshold para dropar a operação

4 - subset: especifica a linha/coluna para valores null







5 - inplace: é um valor booleano, se for True o DataFrame será mudado e o valor None é retornado
